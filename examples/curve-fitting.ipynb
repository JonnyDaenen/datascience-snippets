# Dummy Example

```python
X = np.linspace(0, 100, 50)
Y = 23.24 + 2.2*X + 0.24*(X**2) + 10*np.random.randn(50) #added some noise
```

```python
X
```

```python
Y
```

```python
coefs = np.polyfit(X, Y, 2)
coefs
```

```python
p = np.poly1d(coefs)
p
```

```python
plt.plot(X, p(X), "r-") #p(X) evaluates the polynomial at X
plt.plot(X, Y, "bo", markersize= 2)

plt.show()
```

```python
from numpy.polynomial import Polynomial as P

pp = P.fit(X, Y, 2)
```

```python
plt.plot(X, pp(X), "r-") #p(X) evaluates the polynomial at X
plt.plot(X, Y, "bo", markersize= 2)

plt.show()
```

```python
(pp - 1500).roots()
```

# Shampoo example

```python
import pandas as pd

def parser(x):
    # fake the year numbers
    return pd.datetime.strptime('200'+x, '%Y-%m')

shampoo = pd.read_csv('shampoo-sales.csv', 
                  header=0, # wor #0 is the header row
                  quotechar='"',
                 parse_dates=[0],
                 index_col=0, # date column is index
                 squeeze=True, # return series if only 1 column
                 date_parser=parser)
```

```python
shampoo
```

```python
shampoo.plot()
plt.show()
```

we need x-values and y-values

```python
dataset = shampoo.reset_index()
dataset.head(2)
```

```python
shampoo.values
```

```python
xvals = np.arange(3*12)
coefs = np.polyfit(xvals, shampoo.values, 2)
coefs
```

```python
p = np.poly1d(coefs)
p
```

```python

plt.plot(xvals, shampoo.values, "bo", markersize= 2)
plt.plot(xvals, p(xvals), "r-") #p(X) evaluates the polynomial at X
plt.show()
```

```python
observations = len(shampoo.values)
# 3 years of values
xvals_future = np.arange(observations, observations + 36)
yvals_future = p(xvals_future)
```

```python
markers_on = [12, 17, 18, 60]
# plt.plot(xvals, shampoo.values, "bo", markersize= 2)
plt.plot(xvals, shampoo.values, "c-", markersize= 1)
plt.plot(xvals, p(xvals), "r-") #p(X) evaluates the polynomial at X
plt.plot(xvals_future, yvals_future, "r-", markersize= 4) #  color="red", linestyle="dashed"

plt.show()
```

```python
pp = P.fit(xvals, shampoo.values, 2)
```

```python
roots = (pp - 1500).roots()
result = filter(lambda x: x > 0, roots)
xmarker = list(result)[0]
xmarker = int(np.ceil(xmarker))
```

```python
p
```

```python
markers_on = [12, 17, 18, 60]
# plt.plot(xvals, shampoo.values, "bo", markersize= 2)
plt.plot(xvals, shampoo.values, "c-", markersize= 1)
plt.plot(xvals, p(xvals), "r-") #p(X) evaluates the polynomial at X
plt.plot(xvals_future, yvals_future, "r--", markersize= 4) #  color="red", linestyle="dashed"


plt.plot([xmarker, xmarker], [0, p(xmarker)], "w--", linewidth=1) #  color="red", linestyle="dashed"

plt.plot([xmarker], [p(xmarker)], "ro", markersize=10) #  color="red", linestyle="dashed"
# plt.scatter([xmarker], [p(xmarker)], 50, markersize=10) #  color="red", linestyle="dashed"

# see: https://matplotlib.org/examples/pylab_examples/annotation_demo2.html
plt.annotate(f'limit reached on month {xmarker}',
             xy=(xmarker, p(xmarker)), xycoords='data',
             xytext=(-100, +40), textcoords='offset points', fontsize=12,
             bbox=dict(boxstyle="round4,pad=.5", fc="0.8"),
             arrowprops=dict(shrinkA=10, shrinkB=10, arrowstyle="fancy",  fc='black', ec='black',connectionstyle="arc3,rad=.2"))




plt.show()
```
